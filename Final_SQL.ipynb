```python
# Import libraries
import pandas
import ibm_db
import ibm_db_dbi
dsn_driver   = "{IBM DB2 ODBC DRIVER}"
dsn_database = "BLUDB"
dsn_hostname = ".........bluemix.net"
dsn_port     = "50000"
dsn_protocol = "TCPIP"
dsn_uid      = "Login"
dsn_pwd      = "Password"
# Create DNS connection string
dsn = ("DRIVER     = {0};"
       "DATABASE   = {1};"
       "HOSTNAME   = {2};"
       "PORT       = {3};"
       "PROTOCOL   = {4};"
       "UID        = {5};"
       "PWD        = {6};").format(dsn_driver, dsn_database, dsn_hostname,  dsn_port, dsn_protocol, dsn_uid, dsn_pwd)
```

```python
print(dsn)
```

```python
# Try to connect to ibm DB2 server
try:
    conn = ibm_db.connect(dsn,"","")
    print("Connected!")
except:
    print("No connection...!")
```

```python
# Assigning cursor to variable name pconn
pconn = ibm_db_dbi.Connection(conn)
```

### Problem 2
     Find average college enrollments by community area

```python
selectQuery = '''select "Community_Area_Name",avg("College_Enrollment__number_of_students_") as "averagecollege_enrollments" 
from CHICAGO_PUBLIC_SCHOOLS
group by "Community_Area_Name"'''

pdf = pandas.read_sql(selectQuery,pconn)

pdf
```

### Problem 3
    Find the number of schools that are healthy school certified

```python
selectQuery = """select count("Healthy_Schools_Certified_") as "number_of_healthy_school_certified" 
from CHICAGO_PUBLIC_SCHOOLS 
where "Healthy_Schools_Certified_" in 'Yes'"""

pdf = pandas.read_sql(selectQuery,pconn)

pdf
```

### Problem 4
    How many observations have a Location Description value of GAS STATION

```python
selectQuery = """select count("Location_Description") as "number_of_GAS STATION" from CHICAGO_CRIME_DATA 
where "Location_Description" = 'GAS STATION'"""

pdf = pandas.read_sql(selectQuery,pconn)

pdf
```

### Problem 5
    Retrieve a list of the top 10 community areas which have most number of schools and sorted in descending order.

```python
selectQuery = """select "Community_Area_Name", count("Name_of_School") as "number_of_schools"  from CHICAGO_PUBLIC_SCHOOLS 
group by "Community_Area_Name" order by "number_of_schools" desc limit 10"""

pdf = pandas.read_sql(selectQuery,pconn)

pdf
```

### Problem 6
    How many observations have value MOTOR VEHICLE THEFT in the Primary Type variable (this is the number of crimes related     to Motor vehicles)

```python
selectQuery = """select count("Primary_Type") as "number_MOTOR_VEHICLE_THEFT" from CHICAGO_CRIME_DATA 
where "Primary_Type" = 'MOTOR VEHICLE THEFT'"""

pdf = pandas.read_sql(selectQuery,pconn)

pdf
```

### Problem 7
     Find the minimum “Average Student Attendance” for community are where hardship is 96.

```python
selectQuery = """select S."Average_Student_Attendance",
S."Community_Area_Number",
D."Community_Area_Number",
D."HARDSHIP_INDEX" from CHICAGO_PUBLIC_SCHOOLS S 
INNER JOIN CENSUS_DATA D ON S."Community_Area_Number" = D."Community_Area_Number" 
where D."HARDSHIP_INDEX" = 96 order by S."Average_Student_Attendance" LIMIT 1
"""

pdf = pandas.read_sql(selectQuery,pconn)

pdf
```
